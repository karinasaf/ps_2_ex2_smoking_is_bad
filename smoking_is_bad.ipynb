# Is smoking really good?
Now it is up to you to run the analysis further and check what could be underlying your cousings results.

```python
sns.histplot(data=df, x='age')
```

```python
sns.histplot(data=df, x='age', hue='smokes', bins=20, kde=True)
```

```python
df[['age', 'alive']].corr()
```

```python
df['age_band'] = pd.cut(df['age'], bins=[0,30,50,70,100])
df.groupby(['age_band','smokes'])['alive'].mean()
```

```python
df.groupby('smokes')['age_band'].value_counts()
```

```python
import statsmodels.formula.api as smf
model = smf.logit('alive ~ smokes + age', data=df).fit()
print(model.summary())
```

The results was due to the sample of non-smokers including more people of older age.
